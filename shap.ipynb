```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import shap

model_name = "distilbert-imdb-final"
tokenizer = AutoTokenizer.from_pretrained(model_name, local_files_only=True)
model = AutoModelForSequenceClassification.from_pretrained(model_name, local_files_only=True)
```

```python
from transformers import pipeline
explainer_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
explainer = shap.Explainer(explainer_pipeline)
```

```python
sample_texts = [
    "The film was not good at all.",
    "Absolutely amazing acting and cinematography!",
    "It was okay, but not as bad as people say.",
    "Terrible writing, boring dialogue, awful pacing.",
]
```

```python
shap_values = explainer(sample_texts)
```

```python
tokens = shap_values[0].data
values = shap_values[0].values
```

```python
from transformers import pipeline
import shap

# 1. Load pipeline
clf = pipeline("sentiment-analysis", model="distilbert-imdb-final", tokenizer="distilbert-imdb-final")

# 2. Create text masker
masker = shap.maskers.Text(clf.tokenizer)

# 3. Create explainer
explainer = shap.Explainer(clf, masker)

# 4. Get SHAP values for a text
shap_values = explainer(["This movie was amazing!"])

# 5. Plot as colored text (not waterfall)
shap.plots.text(shap_values[0])
```
